```python
#import most used libraries for machine learning and deep learning
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#read csv as a pandas dataframe
df_train = pd.read_csv('../../data/ptbdb_train.csv')
df_test = pd.read_csv('../../data/ptbdb_test.csv')
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
#separate last column as target
data_matrix = df_train.values
y = data_matrix[:,-1]
X = data_matrix[:,0:-1]

#plot some examples of the data
plt.figure(figsize=(12, 6))
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.plot(X[i-1])
    plt.title(f'Label: {round((y[i-1]))}')
    plt.tight_layout()
plt.show()
```

```python
# Create a histogram of the vector y
plt.figure(figsize=(12, 6))
plt.hist(y, bins='auto', color='blue', edgecolor='black')
plt.title('Histogram of y')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
